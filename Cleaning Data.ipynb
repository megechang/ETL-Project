```python
# Dependencies
import pandas as pd

# Study data files
imdb = "imdb_new.csv"

# Read the csv
imdb_data = pd.read_csv(imdb)
```

```python
# Create a data frame out of the csv data
imdb_data_df = pd.DataFrame(imdb_data)
```

```python
# Show the new data frame
imdb_data
```

```python
# Show clean df
imdb_data_df.head()
```

```python
# Drop any 
imdb_data_df = imdb_data_df.dropna()
```

```python
# Show new table
imdb_data_df.head()
```

```python
# Export the CSV
imdb_data_df.to_csv('IMDB.csv')
```
